```python
def check_type(variable):
  """
  Checks if a variable is a list or a string.

  Args:
    variable: The variable to check.

  Returns:
    "list" if the variable is a list, "string" if it's a string, 
    and "other" otherwise.
  """
  if isinstance(variable, list):
    return "list"
  elif isinstance(variable, str):
    return "string"
  elif isinstance(variable, int):
    return "integer"
  else:
    return "other"

# Examples
my_list = [1, 2, 3]
my_string = "hello"
my_integer = 10

print(f"The type of my_list is: {check_type(my_list)}")
print(f"The type of my_string is: {check_type(my_string)}")
print(f"The type of my_integer is: {check_type(my_integer)}")
```

```python

#buildTable(caption,header,body,footer) -> string
#caption -> string
#header -> list of lists
#body -> list of lists
#footer -> list of lists
#Creates HTML table with caption Caption, using header, body, and footer as such.
#header, body, footer: main list is all data, each inner list is a row, each list value is a collumn for that row
def buildTable(caption,header,body,footer):
            tableMultiline = '''\n\t<table>
            <caption> _CAPTION_ </caption>
            <thead> _HEAD_
            </thead>
            <tbody> _BODY_
            </tbody>
            <tfoot> _FOOT_
            </tfoot>
        </table>\n'''
            multHead = ""
            multBody = ""
            multFoot = ""
            for item in header:
                multHead += "\n\t\t<tr>"
                for cell in item:
                    multHead += "<th>" + str(cell) + "</th>"
                multHead += "</tr>"
            for item in body:
                multBody += "\n\t\t<tr>"
                for cell in item:
                    multBody += "<td>" + str(cell) + "</td>"
                multBody += "</tr>"
            for item in footer:
                multFoot += "\n\t\t<tr>"
                for cell in item:
                    multFoot += "<td>" + str(cell) + "</td>"
                multFoot += "</tr>"
            tableMultiline = tableMultiline.replace("_CAPTION_",caption)
            tableMultiline = tableMultiline.replace("_HEAD_",multHead)
            tableMultiline = tableMultiline.replace("_BODY_",multBody)
            tableMultiline = tableMultiline.replace("_FOOT_",multFoot)
            return tableMultiline
```

```python

with open("pokemon/pokemon.csv", "r") as pokeData:
    pokeData = pokeData.read()
    pokeData = pokeData.split("\n")
    pokeDataSplit = []
    for item in pokeData:
        pokeDataSplit.append(item.split(","))
typesList = ["bug","dragon","electric","fairy","fighting","fire","flying","ghost","grass","ground","ice","normal","poison","psychic","rock","steel","water"]
grassList = [pokemon for pokemon in pokeDataSplit if "grass" in [type_.lower() for type_ in pokemon]]
bigTypesListForTable = []
for item in typesList:
    bigTypesListForTable.append([pokemon for pokemon in pokeDataSplit if f"{item}" in [type_.lower() for type_ in pokemon]])
print(bigTypesListForTable)


for i in range(len(typesList)):
    print(buildTable(f"{typesList[i]}",[pokeDataSplit[0]],[bigTypesListForTable[i]],[]))
```

```python
def readURL(url = "", decodeType = 'utf-8'):
    import urllib.request
    if not(url):
        raise ValueError("Invalid URL: Please provide a URL.")
    try:
        urllib.request.urlopen(url)
    except:
        raise ValueError("Invalid URL: Please provide a valid URL. Try adding 'https://' or 'http://' at the start.")
    return urllib.request.urlopen(url).read().decode(decodeType)
print(readURL("https://raw.githubusercontent.com/kccaterworld/kcater70IntroToCS/refs/heads/main/Python/Misc/turtcolors.txt").split("\n"))
print(readURL("james.com"))
```
